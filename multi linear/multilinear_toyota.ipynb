```python
import statsmodels.formula.api as smf #for regression model
```

```python
ml1=smf.ols('Price~Age_08_04+KM+HP+Cylinders+Gears', data=price_car).fit() # regression model
```

```python
ml1.params
```

```python
ml1.summary()
```
